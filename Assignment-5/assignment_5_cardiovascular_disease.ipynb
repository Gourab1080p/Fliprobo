```python
#As we can see weight and height seem to be unrelated to cadio but body mass index feature may be helpful
df["bmi"] = df["weight"] / (df["height"]/100)**2
```

```python
#removing outliers
df.drop(df[(df['ap_hi'] > df['ap_hi'].quantile(0.975)) | (df['ap_hi'] < df['ap_hi'].quantile(0.025))].index,inplace=True)
df.drop(df[(df['ap_lo'] > df['ap_lo'].quantile(0.975)) | (df['ap_lo'] < df['ap_lo'].quantile(0.025))].index,inplace=True)
```

```python
df.head(5)
```

```python
df.shape
```

```python
df.drop(['height','weight'], axis=1,inplace=True)
```

```python
#splitting data into training and testing 
X = df.drop(['cardio'], axis=1)
y = df['cardio']

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

# Selecting model

1. Support Vector Machine

```python
#Support Vector Machine
from sklearn.svm import SVC
svc = SVC(gamma='auto')
svc.fit(X_train, y_train)
```

```python
y_pred_svc = svc.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
print(confusion_matrix(y_test, y_pred_svc))
print(classification_report(y_test, y_pred_svc))
```

```python
svc.score(X_test, y_test)
```

```python
#Applying k-Fold Cross Validation
from sklearn.model_selection import cross_val_score
acc_svc = cross_val_score(estimator=svc, X=X_train, y=y_train, cv=10, n_jobs=4)
acc_svc
```

```python
print(acc_svc.mean())
print(acc_svc.std())
```

2. Random forest model

```python
#Random Forest model
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(n_estimators=100)
rfc.fit(X_train, y_train)
y_pred_rfc = rfc.predict(X_test)
```

```python
# Random Forest Model Evaluation
print(confusion_matrix(y_test, y_pred_rfc))
print(classification_report(y_test, y_pred_rfc))
```

```python
rfc.score(X_test, y_test)
```

```python
acc_rfc = cross_val_score(estimator=rfc, X=X_train, y=y_train, cv=10)
acc_rfc
```

```python
print(acc_rfc.mean())
print(acc_rfc.std())
```

3. Naive Bayes model

```python
# Naive Bayes
from sklearn.naive_bayes import GaussianNB
nbc = GaussianNB()
nbc.fit(X_train, y_train)
y_pred_nbc = nbc.predict(X_test)
```

```python
print(confusion_matrix(y_test, y_pred_nbc))
print(classification_report(y_test, y_pred_nbc))
```

```python
nbc.score(X_test, y_test)
```

```python
#Applying k-Fold Cross Validation
acc_nbc = cross_val_score(estimator=nbc, X=X_train, y=y_train, cv=10)
acc_nbc
```

```python
print(acc_nbc.mean())
print(acc_nbc.std())
```

4. XGBoost Model

```python
# XGBoost Model
from xgboost import XGBClassifier
xgb = XGBClassifier(learning_rate=0.02, n_estimators=600)
xgb.fit(X_train, y_train)
```

```python
y_pred_xgb = xgb.predict(X_test)
```

```python
# XGBoost Model Evaluation
print(confusion_matrix(y_test, y_pred_xgb))
print(classification_report(y_test, y_pred_xgb))
```

```python
xgb.score(X_test, y_test)
```

```python
#Applying k-Fold Cross Validation
acc_xgb = cross_val_score(estimator=xgb, X=X_train, y=y_train, cv=10)
acc_xgb
```

```python
print(acc_xgb.mean())
print(acc_xgb.std())
```

5. K Nearest Neighbor Model

```python
# KNN
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=100)
knn.fit(X_train, y_train)
```

```python
y_pred_knn = knn.predict(X_test)
```

```python
# KNN Model Evaluation
print(confusion_matrix(y_test, y_pred_knn))
print(classification_report(y_test, y_pred_knn))
```

```python
knn.score(X_test, y_test)
```

```python
#Applying k-Fold Cross Validation
acc_knn = cross_val_score(estimator=knn, X=X_train, y=y_train, cv=10)
acc_knn
```

```python
print(acc_knn.mean())
print(acc_knn.std())
```

The XGBoost model was the best performer out of the five models giving us a mean accuracy score of 73%
